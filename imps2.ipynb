### **3. How Convolution Works**

A **3√ó3 filter** slides across the image like a stamp:
- At each position: multiply filter values with image pixels
- Sum all products = single output value
- Repeat for entire image = output feature map

**Example**: 
- Filter detects vertical edges ‚Üí highlights vertical lines in image
- 32 filters = 32 different edge/pattern detectors
- Each filter learns different features automatically

### **5. Training Process**

**Epoch** = one complete pass through all training data

**Each epoch:**
1. Feed 32 images (batch) through model
2. Calculate loss (error between prediction & truth)
3. Update weights using Adam optimizer (backpropagation)
4. Repeat for all batches
5. Validate on validation set
6. Check callbacks (stop early? reduce learning rate?)

**Callbacks:**
- **EarlyStopping**: Stop if validation loss doesn't improve for 4 epochs
- **ReduceLROnPlateau**: If stuck, multiply learning rate by 0.5 (take smaller steps)
